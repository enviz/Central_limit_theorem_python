```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def flip_coin(num):
    

    L1 = [random.randint(0,1) for r in range(num)]
    
    return L1



def plot_histogram(N,n_samples,n_tosses):    
    #N = 1000                         #user_defined
    #n_samples = 20
    #n_tosses = 800
    results = flip_coin(n_tosses)
    L2 = []

    for i in range(N):
        L2.append(random.sample(results, n_samples))
    mean_list = np.mean(L2, axis=1)
    print('mean values are:',mean_list)



    plt.hist(mean_list, bins=10)
    plt.show()
```

```python
plot_histogram(1000,20,800)
```

```python
for i in range(10):

    plot_histogram(1000,20,800)
```
